# Import Library

```python
import csv
import pandas as pd
```

# Load 3 CSV files

```python
train = pd.read_csv(
    r"D:\CSE DEPT DOCUMENT\11th Semester\THESIS\DATASET\Training.csv",
    usecols=["id", "acts", "emotions", "utterances", "context_state"]
)

test = pd.read_csv(
    r"D:\CSE DEPT DOCUMENT\11th Semester\THESIS\DATASET\Testing.csv",
    usecols=["id", "acts", "emotions", "utterances", "context_state"]
)

val = pd.read_csv(
    r"D:\CSE DEPT DOCUMENT\11th Semester\THESIS\DATASET\Validation.csv",
    usecols=["id", "acts", "emotions", "utterances", "context_state"]
)


print("Loaded → Train:", len(train))
print("Loaded → Test:", len(test))
print("Loaded → Validation:", len(val))
```

# Preprocessing function

```python
def preprocess(df: pd.DataFrame, name: str) -> pd.DataFrame:
    original_len = len(df)
 
    # 2a. Remove exact duplicate rows
    df = df.drop_duplicates()
    after_dup = len(df)
 
    # 2b. Drop rows where any column EXCEPT utterances is null or empty string
    cols_to_check = ["id", "acts", "emotions", "context_state"]
    df = df.dropna(subset=cols_to_check)
    for col in cols_to_check:
        df = df[df[col].astype(str).str.strip() != ""]
    after_null = len(df)
 
    # 2c. Strip leading/trailing whitespace from id and context_state only
    df["id"]            = df["id"].astype(str).str.strip()
    df["context_state"] = df["context_state"].astype(str).str.strip()
 
    # 2d. acts and emotions: keep as-is (numeric list strings), just strip outer spaces
    df["acts"]     = df["acts"].astype(str).str.strip()
    df["emotions"] = df["emotions"].astype(str).str.strip()

 
    # 2f. Reset index
    df = df.reset_index(drop=True)
 
    print(f"\n[{name}]")
    print(f"  Original rows  : {original_len}")
    print(f"  After dedup    : {after_dup}  (removed {original_len - after_dup} duplicates)")
    print(f"  After null-drop: {after_null}  (removed {after_dup - after_null} null rows)")
    print(f"  Final rows     : {len(df)}")
 
    return df
```

# Apply

```python
train_clean = preprocess(train, "Train")
test_clean  = preprocess(test,  "Test")
val_clean   = preprocess(val,   "Validation")
```

# Save Data

```python
out_train = r"D:\CSE DEPT DOCUMENT\11th Semester\THESIS\CLEANED DATASET\Training_clean.csv"
out_test  = r"D:\CSE DEPT DOCUMENT\11th Semester\THESIS\CLEANED DATASET\Testing_clean.csv"
out_val   = r"D:\CSE DEPT DOCUMENT\11th Semester\THESIS\CLEANED DATASET\Validation_clean.csv"
```

# QUOTE_ALL ensures utterances (with internal commas/semicolons/quotes) stay intact

```python
train_clean.to_csv(out_train, index=False, quoting=csv.QUOTE_ALL)
test_clean.to_csv(out_test,   index=False, quoting=csv.QUOTE_ALL)
val_clean.to_csv(out_val,     index=False, quoting=csv.QUOTE_ALL)
 
print("\n✓ Clean files saved:")
print(" ", out_train)
print(" ", out_test)
print(" ", out_val)
```
