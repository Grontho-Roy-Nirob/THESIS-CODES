```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Dataset from HuggingFace parquet URL 
train_df = pd.read_parquet("https://huggingface.co/datasets/roskoN/dailydialog/resolve/refs%2Fconvert%2Fparquet/full/train/0000.parquet")
val_df = pd.read_parquet("https://huggingface.co/datasets/roskoN/dailydialog/resolve/refs%2Fconvert%2Fparquet/full/validation/0000.parquet")
test_df = pd.read_parquet("https://huggingface.co/datasets/roskoN/dailydialog/resolve/refs%2Fconvert%2Fparquet/full/test/0000.parquet")

print("Train shape:", train_df.shape)
print("Val shape:", val_df.shape)
print("Test shape:", test_df.shape)

print("\n--- Columns ---")
print(train_df.columns.tolist())

print("\n--- Sample Row ---")
print(train_df.iloc[0])
```

```python
print("=" * 40)
print("Dataset Split Summary")
print("=" * 40)
print(f"Train      : {len(train_df)} rows")
print(f"Validation : {len(val_df)} rows")
print(f"Test       : {len(test_df)} rows")
print(f"Total      : {len(train_df) + len(val_df) + len(test_df)} rows")
print("=" * 40)

# Percentage
total = len(train_df) + len(val_df) + len(test_df)
print(f"\nTrain      : {len(train_df)/total*100:.1f}%")
print(f"Validation : {len(val_df)/total*100:.1f}%")
print(f"Test       : {len(test_df)/total*100:.1f}%")
```

```python
print("=" * 50)
print("TRAIN - First 5 Rows")
print("=" * 50)
for i in range(5):
    print(f"\n[{i+1}] ID: {train_df.iloc[i]['id'][:20]}...")
    print(f"    Utterances : {train_df.iloc[i]['utterances']}")
    print(f"    Emotions   : {train_df.iloc[i]['emotions']}")
    print(f"    Acts       : {train_df.iloc[i]['acts']}")

print("\n" + "=" * 50)
print("VALIDATION - First 5 Rows")
print("=" * 50)
for i in range(5):
    print(f"\n[{i+1}] ID: {val_df.iloc[i]['id'][:20]}...")
    print(f"    Utterances : {val_df.iloc[i]['utterances']}")
    print(f"    Emotions   : {val_df.iloc[i]['emotions']}")
    print(f"    Acts       : {val_df.iloc[i]['acts']}")

print("\n" + "=" * 50)
print("TEST - First 5 Rows")
print("=" * 50)
for i in range(5):
    print(f"\n[{i+1}] ID: {test_df.iloc[i]['id'][:20]}...")
    print(f"    Utterances : {test_df.iloc[i]['utterances']}")
    print(f"    Emotions   : {test_df.iloc[i]['emotions']}")
    print(f"    Acts       : {test_df.iloc[i]['acts']}")
```
